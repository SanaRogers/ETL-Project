```python
import pandas as pd
from sqlalchemy import create_engine
```

### Store CSV into DataFrame

```python
csv_file = "Resources/2017_2018_Stats.csv"
stats_df = pd.read_csv(csv_file)
stats_df.head()
```

```python
csv_file = "Resources/2017_2018_Salary.csv"
salary_df = pd.read_csv(csv_file)
salary_df.head()
```

```python
csv_file = "Resources/all_players.csv"
all_players_df = pd.read_csv(csv_file)
all_players_df.head()
```

```python
print(stats_df.columns)
```

```python
print(salary_df.columns)
```

```python
print(all_players_df.columns)
```

```python
pl_grp = all_players_df.groupby(['player_name','team_abbreviation'])
players_df = pl_grp.mean().reset_index().drop(columns='Unnamed: 0').rename(columns={'player_name':'players','team_abbreviation':'teams'})
```

```python
players_df.head()
```

```python
salary_df.columns
```

```python
s_grp = salary_df.groupby(['Player','Tm'])
```

```python
sal_df = s_grp.mean().drop(columns = 'Unnamed: 0').rename(columns={'Player': 'players', 'Tm':'teams'}).reset_index()
```

```python
stats_df.describe().columns
```

```python
stats_grp = stats_df.groupby(['Player','Tm'])
```

```python
st_df = stats_grp.mean()[['Rk', 'Age', 'G', 'MP', 'PER', 'TS%', '3PAr', 'FTr', 'ORB%', 'DRB%','TRB%', 'AST%', 'STL%', 'BLK%', 'TOV%', 'USG%','OWS','DWS', 'WS', 'WS/48', 'OBPM', 'DBPM', 'BPM', 'VORP']].reset_index().rename(columns={'Player':'players','Tm':'teams'})
st_df.head()
```

```python
st_df["players"]=st_df["players"].str.split("\\",n=1, expand=True)
st_df.head(100)


                            
```

```python
players_df.to_csv('Resources/players.csv')
sal_df.to_csv('Resources/salary.csv')
st_df.to_csv('Resources/stats.csv')
```
