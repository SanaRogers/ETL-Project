```python
import pandas as pd
from sqlalchemy import create_engine
```

### Store 3 CSV files into DataFrames

```python
csv_file = "Resources/2017_2018_Stats.csv"
stats_df = pd.read_csv(csv_file)
stats_df.head()
```

```python
csv_file = "Resources/2017_2018_Salary.csv"
salary_df = pd.read_csv(csv_file)
salary_df.head()
```

```python
csv_file = "Resources/all_players.csv"
draft_df = pd.read_csv(csv_file)
draft_df.head()
```

```python
print(stats_df.columns)
```

```python
print(salary_df.columns)
```

```python
print(draft_df.columns)
```

```python
# draft_final_df = draft_df.groupby(['player_name','team_abbreviation'])
# draft_info_df = draft_final_df.mean().reset_index().drop(columns='Unnamed: 0').rename(columns={'player_name':'Player','team_abbreviation':'Team'})
draft_df = draft_df.rename(columns={'player_name':'Player', 'team_abbreviation':'Team'})
draft_final_df = draft_df[['Player','Team','draft_year','draft_round','draft_number']]
draft_final_df.head()
```

```python
salary_df.columns
```

```python
s_grp = salary_df.groupby(['Player','Tm'])
```

```python
sal_df = s_grp.mean().drop(columns = 'Unnamed: 0').rename(columns={'Player': 'players', 'Tm':'teams'}).reset_index()
```

```python
stats_df.describe().columns
```

```python
salary_final_df = sal_df.rename(columns = {'season17_18':'season17_18_salary'})
salary_final_df.head()
```

```python
stats_grp = stats_df.groupby(['Player','Tm'])
```

```python
st_df = stats_grp.mean()[['Rk', 'Age', 'G', 'MP', 'PER', 'TS%', '3PAr', 'FTr', 'ORB%', 'DRB%','TRB%', 'AST%', 'STL%', 'BLK%', 'TOV%', 'USG%','OWS','DWS', 'WS', 'WS/48', 'OBPM', 'DBPM', 'BPM', 'VORP']].reset_index().rename(columns={'Player':'players','Tm':'teams'})
st_df.head()
```

```python
st_df["players"]=st_df["players"].str.split("\\",n=1, expand=True)
st_df.head(100)


                            
```

```python
per_info_df = st_df.rename(columns = {'players':'Player'})
per_info_final_df = per_info_df[['Player','teams','PER']]
per_info_final_df.head()
```

# Export Updated CSV files

```python
draft_final_df.to_csv('Resources/DraftInfo.csv', index=False)
salary_final_df.to_csv('Resources/Salaries.csv', index=False)
per_info_final_df.to_csv('Resources/PerInfo.csv', index=False)
```
